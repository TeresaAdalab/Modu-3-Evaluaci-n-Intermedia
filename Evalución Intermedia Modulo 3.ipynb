```python
# Crear una lista de tuplas para iterar sobre los DataFrames y sus nombres
dataframes = [
    ('compilation', df_compilation),
    ('single', df_single),
    ('album', df_album)
]
```

```python
# Lista de columnas de interés
columnas_interes = ['danceability', 'acousticness', 'speechiness']
```

```python
# Crear un diccionario para almacenar los resultados finales
resultados_por_tipo = {}
```

```python
# Iterar sobre cada DataFrame y calcular la media y desviación estándar agrupadas por artista
for tipo, df in dataframes:
    # Agrupar por artista y calcular la media y la desviación estándar
    resultados = df.groupby('artist')[columnas_interes].agg(['mean', 'std'])
    
    # Ordenar por la media de 'danceability' en orden descendente
    resultados_sorted = resultados.sort_values(by=('danceability', 'mean'), ascending=False)
    
    # Agregar los resultados al diccionario usando el tipo de álbum como clave
    resultados_por_tipo[tipo] = resultados_sorted

# Mostrar los resultados ordenados
for tipo, resultados in resultados_por_tipo.items():
    print(f"\nResultados ordenados para tipo de álbum '{tipo}' por 'danceability':")
    print(resultados)
```

```python
# Agrupar por artistaa y calcular la media de visitas y likes
resultados_artista = df.groupby('artist')[['views', 'likes']].mean()
resultados_artista
```

```python

# Agrupar por artista y sumar las visitas
visitas_por_artista = df.groupby('artist')['views'].sum()

# Calcular el total de visitas de todos los artistas
total_visitas = visitas_por_artista.sum()

# Calcular el porcentaje de visitas para cada artista
visitas_porcentaje = (visitas_por_artista / total_visitas) * 100

# Ordenar los artistas por visitas en orden descendente y quedarse con los primeros 10
top_artistas_por_visitas = visitas_porcentaje.sort_values(ascending=False).head(10)

# Mostrar los 10 artistas con más visitas en porcentaje
print("Top 10 artistas por visitas en porcentaje:")
print(top_artistas_por_visitas)
```
