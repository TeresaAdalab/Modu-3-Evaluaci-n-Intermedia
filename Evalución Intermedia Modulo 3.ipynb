```python
# 2. Preparación de Datos
## 2.1 Columnas en minuscula
# Convertir nombres de columnas a minúsculas
df.columns = df.columns.str.lower()

# Mostrar los nombres de las columnas actualizados
print(df.columns)
```

```python
columnas_a_actualizar = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
```

```python
# Función para reemplazar comas por puntos en todas las columnas
def reemplazar_comas_por_puntos_en_columnas(df, columnas):
    for columna in columnas:
        if df[columna].dtype == 'object':
            df[columna] = df[columna].str.replace(',', '.').astype(float)
    return df
```

```python
df = reemplazar_comas_por_puntos_en_columnas(df, columnas_a_actualizar)
```

```python
def convertir_a_minusculas(df):
    # Convertir títulos de canción (Track) a minúsculas
    df['track'] = df['track'].str.lower()
    
    # Convertir títulos de vídeos (title) a minúsculas
    df['title'] = df['title'].str.lower()
    
    # Convertir nombres de álbum (album) a minúsculas
    df['album'] = df['album'].str.lower()
    
    return df
```

```python
def reemplazar_simbolo_dolar(df):
    # Reemplazar '$' por 's' en la columna 'artist'
    df['artist'] = df['artist'].str.replace('$', 's')
    
    return df
```

```python
# Crear una lista de tuplas para iterar sobre los DataFrames y sus nombres
dataframes = [
    ('compilation', df_compilation),
    ('single', df_single),
    ('album', df_album)
]
```

```python
# Lista de columnas de interés
columnas_interes = ['danceability', 'acousticness', 'speechiness']
```

```python
# Crear un diccionario para almacenar los resultados finales
resultados_por_tipo = {}
```

```python
# Iterar sobre cada DataFrame y calcular la media y desviación estándar agrupadas por artista
for tipo, df in dataframes:
    # Agrupar por artista y calcular la media y la desviación estándar
    resultados = df.groupby('artist')[columnas_interes].agg(['mean', 'std'])
    
    # Ordenar por la media de 'danceability' en orden descendente
    resultados_sorted = resultados.sort_values(by=('danceability', 'mean'), ascending=False)
    
    # Agregar los resultados al diccionario usando el tipo de álbum como clave
    resultados_por_tipo[tipo] = resultados_sorted

# Mostrar los resultados ordenados
for tipo, resultados in resultados_por_tipo.items():
    print(f"\nResultados ordenados para tipo de álbum '{tipo}' por 'danceability':")
    print(resultados)
```
